# Lab 2 - Scale Invariant Feature Transform

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
eiffel_img = cv2.imread("img/Eiffel_Tower.jpg")
plt.imshow(eiffel_img)
```

```python
gray = cv2.cvtColor(eiffel_img, cv2.COLOR_BGR2GRAY)
sift = cv2.xfeatures2d.SIFT_create()
kp = sift.detect(gray, None)
img = cv2.drawKeypoints(gray, kp, eiffel_img)
```
